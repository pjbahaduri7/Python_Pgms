<p>1. We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically using the scipy module, the dataset given below.</p>
<p>Task:<br>
1.fitting it to the periodic function<br>
2.plot the fit</p>
<p>Data<br>
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25<br>
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
months = np.arange(12)
```

```python
def yearly_temp(times, avg, ampl, time_offset):
    return (avg+ampl*np.cos((times+time_offset)*2*np.pi/times.max()))
res_max, cov_max = optimize.curve_fit(yearly_temp, months, temp_max, [20,10,0])
res_min, cov_min = optimize.curve_fit(yearly_temp, months, temp_min, [-40,20,0])
```

```python
days = np.linspace(0, 12, num=365)
plt.figure()
plt.plot(months, temp_max,'ro')
plt.plot(days, yearly_temp(days, *res_max), 'r-')
plt.plot(months, temp_min,'bo')
plt.plot(days, yearly_temp(days, *res_min), 'b-')
plt.xlabel("Months")
plt.ylabel("Temperature ($^\circ$C)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import warnings as w
w.filterwarnings('ignore')
```

```python
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic
```

```python
# titanic = titanic.iloc[:-1]
titanic
```

```python
titanic['sex'].unique()
```

```python
df = titanic.groupby('sex')['pclass'].count()
df
```

```python
df.plot.pie(label=('sex'), title = "Male/Female Proportions",colormap='autumn', autopct='%1.0f%%')
plt.show()
```

```python
titanic['age'] = titanic['age'].replace({'NaN':np.nan}).astype(float)
titanic
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

colors = {'male':'tab:green', 'female':'tab:pink'}

ax.scatter(titanic['fare'], titanic['age'], c=titanic['sex'].map(colors))

handles = [Line2D([0], [0], marker='o', color='w', markerfacecolor=v, label=k, markersize=8) for k, v in colors.items()]
ax.legend(title='sex', handles=handles, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel("Fares Paid")
plt.ylabel("Age")
plt.show()
```
