```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets as d
```

```python
iris = d.load_iris()
```

```python
X_red = PCA(n_components = 3).fit_transform(iris.data)
Y = iris.target
fig = plt.figure(1,figsize=(8,7))
ax = Axes3D(fig, azim=-110, elev=20)
ax.scatter(X_red[:,0], X_red[:,1], X_red[:,2], c=Y, marker="*", s=50,cmap="prism", edgecolor="black")
ax.set_title("3D chart with transformed dimensions of Iris data")
ax.set_xlabel("First Eigenvector")
ax.set_ylabel("Second Eigenvector")
ax.set_zlabel("Third Eigenvector")
plt.show()
```
